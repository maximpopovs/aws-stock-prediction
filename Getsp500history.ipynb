```python
import yfinance as yf
import boto3 as bt
import pandas as pd
```

```python
##^GSPC, MSFT, KO, RSPU, F, WDC,
symbol= "^GSPC"
sp500 = yf.Ticker(symbol)
data=sp500.history(period="max")
ticklers = []
ticklers = [symbol] * len(data)
data.loc[:,"tickler"]=ticklers
if(symbol=="^GSPC"):
  symbol="SP500"
data.to_csv("{0}-hist.csv".format(symbol))
data.head()
```
